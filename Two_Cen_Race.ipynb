## Filter Data

```python
df2 = df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020)]
```

```python
df2 #filtered data
```

```python
df2["Athlete performance"] = df2["Athlete performance"].str.replace("h"," ")
```

```python
df2
```

```python
df2["Age"] = df["Year of event"] - df["Athlete year of birth"]
```

```python
df2["Age"].dropna()
```

```python
df2["Age"].replace(np.inf,0)
```

```python
df2["Age"].fillna(0)
```

```python
df2= df2.drop(["Athlete club","Athlete country","Athlete age category","Athlete year of birth"], axis = 1)
```

## Remove Null Values

```python
df2.isnull().sum()
```

## Remove Duplicated

```python
df2[df2.duplicated()== True] # No Dups
```

## Fix dtypes

```python
df2.dtypes
```

```python
df2['Age'] = df2['Age'].replace([np.inf, -np.inf], np.nan)
df2['Age'] = df2['Age'].fillna(0)
df2['Age'] = df2['Age'].astype(int)
```

```python
df2.dtypes
```

```python
df2 = df2.rename(columns = { "Year of event":"Year",
                            "Event dates":"E_Date",
                            "Event name" : "E_name",
                            "Event distance/length": "E_Distance",
                            "Event number of finishers":"E_num_of_Finishers",
                            "Athlete performance":"Athlete_performance",
                            "Athlete gender":"Athlete_gender",
                            "Athlete average speed" : "Athlete_average_speed",
                            "Athlete ID":"Athlete_ID",
                            "Age":"Age"
                            })
```
